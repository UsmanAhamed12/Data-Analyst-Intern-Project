# World cup players data cleaning

```python
## import libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline
pd.set_option('display.max_columns',None)
```

```python
## import plauyers dataset

Data1 = pd.read_csv('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/FIFA WORLD CUP ANALYSIS PROJECT/Data/WorldCupPlayers.csv')
Data1.head()
```

```python
## see shape of the dataset
Data1.shape
```

```python
### identify the total duplicate rows in this dataset

Data1.duplicated().sum()
```

```python
## Remove Duplicate rows
Data1.drop_duplicates(inplace= True)
```

```python
### identify the total NAN values in this dataset
Data1.isnull().sum()
```

```python
Data1.info()
```

```python
## import world cup match dataset

Data2 = pd.read_csv('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/FIFA WORLD CUP ANALYSIS PROJECT/Data/cleaned_worldcupmatches.csv')
Data2.head()
```

```python
#### merge two datasets

Data = pd.merge(Data1, Data2, how= 'left', on=['RoundID', 'MatchID'])
Data.head()
```

```python
Data.isnull().sum()
```

```python
Data['Position'].fillna('Unknown', inplace= True)
```

```python
Data.head()
```

```python
# Drop Event column
Data.drop('Event', axis= 1, inplace= True)
```

```python
Data.to_csv('New_Worldcupplayers.csv', index= False)
```

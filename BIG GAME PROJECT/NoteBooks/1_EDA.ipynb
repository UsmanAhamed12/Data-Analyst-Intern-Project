```python
### import libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns',None)
```

```python
big_game_data = pd.read_excel('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BIG GAME PROJECT/Data/Cleaned BIG GAME Census data.xlsx')
all_places_data = pd.read_excel('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BIG GAME PROJECT/Data/All Places Census 2016 Population Estimates.xlsx')
```

```python
#### import Dataset

big_game_data['Player Birthplace (Combo)'] = big_game_data['Player Birthplace (Combo)'].str.strip().str.lower()
all_places_data['Geography, full name (City, State)'] = all_places_data['Geography, full name (City, State)'].str.strip().str.lower()
```

```python
merged_data = big_game_data.merge(
    all_places_data,
    left_on='Player Birthplace (Combo)',
    right_on='Geography, full name (City, State)',
    how='left'
)
```

```python
merged_data.head()
merged_data.isnull().sum()  # Check for unmatched rows
```

```python
unmatched_rows = merged_data[merged_data['Geography, full name (City, State)'].isnull()]
print(unmatched_rows[['Player Name', 'Player Birthplace (Combo)']])
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

pd.set_option('display.max_columns',None)
```

```python
# Load datasets
big_game_df = pd.read_excel('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BIG GAME PROJECT/Data/Cleaned BIG GAME Census data.xlsx')
places_population_df = pd.read_excel('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BIG GAME PROJECT/Data/All Places Census 2016 Population Estimates.xlsx')
```

```python
# Split Geography, full name (City, State) into City and State columns
places_population_df[['City', 'State']] = places_population_df['Geography, full name (City, State)'].str.split(', ', expand=True)
```

```python
# Merge on City and State fields
merged_df = pd.merge(
    big_game_df,
    places_population_df,
    left_on=['Player Birthplace (city, town, etc.)', 'Player Birth State'],
    right_on=['City', 'State'],
    how='left'
)
```

```python
# Load the state population data
state_population_df = pd.read_excel('/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BIG GAME PROJECT/Data/All states Census 2017 Population Estimates.xlsx')
```

```python
# Remove whitespace or standardize state names (optional)
state_population_df['Geography Name'] = state_population_df['Geography Name'].str.strip()
merged_df['Player Birth State'] = merged_df['Player Birth State'].str.strip()
```

```python
# Merge based on state columns
final_merged_df = pd.merge(
    merged_df,
    state_population_df,
    left_on='Player Birth State',  # or 'State GEO ID' if available
    right_on='Geography Name',  # or 'GEOID' if using GEO ID columns
    how='left'
)
```

```python
final_merged_df.head()
```

```python
# Count players by state
state_counts = final_merged_df['Player Birth State'].value_counts().sort_values(ascending= False)
print(state_counts.head(10))  # Display top 10 states with the highest representation

# Combine city and state to get unique city-state combinations
final_merged_df['City_State'] = final_merged_df['Player Birthplace (city, town, etc.)'] + ', ' + final_merged_df['Player Birth State']

# Count players by city
city_counts = final_merged_df['City_State'].value_counts().sort_values(ascending= False)
print(city_counts.head(10))  # Display top 10 cities with the highest representation
```

```python
# City-level population analysis
city_population_avg = final_merged_df['Population Estimate (as of July 1) - 2016_x'].mean()
city_population_median = final_merged_df['Population Estimate (as of July 1) - 2016_x'].median()

print("Average city population for player birthplaces:", city_population_avg)
print("Median city population for player birthplaces:", city_population_median)

# State-level population analysis
state_population_avg = final_merged_df['Population Estimate (as of July 1) - 2017'].mean()
state_population_median = final_merged_df['Population Estimate (as of July 1) - 2017'].median()

print("Average state population for player birthplaces:", state_population_avg)
print("Median state population for player birthplaces:", state_population_median)
```

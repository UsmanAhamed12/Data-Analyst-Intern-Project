```python
## find the NULL Values in this data set 

Data.isnull().sum()
```

```python
### create list only NAN Values features

feature_nan = [feature for feature in Data.columns if Data[feature].isnull().sum() > 0]
feature_nan
```

#### ['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 17', 'Unnamed: 19', 'Unnamed: 22'] in this columns all are NAN Values so let's dropn this columns

```python
## drop NAN value Columns 

Data.drop(feature_nan, axis=1, inplace=True)
```

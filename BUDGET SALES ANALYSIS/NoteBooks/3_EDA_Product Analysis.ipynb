# Product Analysis

```python
### import libraries 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

pd.set_option('display.max_columns',None)
%matplotlib inline
```

## What are the most and least sold domain categories

```python
# Load datasets
sales_path = "/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BUDGET SALES ANALYSIS/Data/Sales.xlsx"
product_path = "/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BUDGET SALES ANALYSIS/Data/Product.xlsx"

sales_df = pd.read_excel(sales_path)
product_df = pd.read_excel(product_path)

# Merge Sales and Product datasets on ProductKey
merged_data = pd.merge(sales_df, product_df, on='ProductKey', how='inner')

# Group by Category and sum SalesAmount
category_sales = merged_data.groupby('Category')['SalesAmount'].sum().reset_index()

# Sort by SalesAmount
category_sales_sorted = category_sales.sort_values(by='SalesAmount', ascending=False)

# Most sold category
most_sold = category_sales_sorted.iloc[0]

# Least sold category
least_sold = category_sales_sorted.iloc[-1]

# Display results
print("Most Sold Category:")
print(most_sold)

print("\nLeast Sold Category:")
print(least_sold)
```

```python
# Bar Chart of Sales by Category
plt.figure(figsize=(10, 6))
plt.bar(category_sales_sorted['Category'], category_sales_sorted['SalesAmount'], color='skyblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

## What are the average prices for domains in different categories

```python
# Load Product dataset
product_path = "/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BUDGET SALES ANALYSIS/Data/Product.xlsx"
product_df = pd.read_excel(product_path)

# Check for relevant columns
print(product_df.columns)

# Group by Category and calculate average ListPrice
category_avg_price = product_df.groupby('Category')['ListPrice'].mean().reset_index().sort_values(by='ListPrice', ascending=False)

# Group by SubCategory and calculate average ListPrice
subcategory_avg_price = product_df.groupby('SubCategory')['ListPrice'].mean().reset_index().sort_values(by='ListPrice', ascending=False)

# Display results
print("Average Prices by Category:")
print(category_avg_price)

print("\nAverage Prices by SubCategory:")
print(subcategory_avg_price)
```

```python
# Bar chart for categories
plt.figure(figsize=(10, 6))
plt.bar(category_avg_price['Category'], category_avg_price['ListPrice'], color='skyblue')
plt.title('Average Prices by Category')
plt.xlabel('Category')
plt.ylabel('Average List Price')
plt.xticks(rotation=45)
plt.show()

# Bar chart for subcategories
plt.figure(figsize=(12, 6))
plt.bar(subcategory_avg_price['SubCategory'], subcategory_avg_price['ListPrice'], color='lightgreen')
plt.title('Average Prices by SubCategory')
plt.xlabel('SubCategory')
plt.ylabel('Average List Price')
plt.xticks(rotation=90)
plt.show()
```

## Are there trends in domain sales based on product attributes

```python
# Load datasets
sales_path = "/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BUDGET SALES ANALYSIS/Data/Sales.xlsx"
product_path = "/Users/inazawaelectronics/Desktop/Data-Analyst-Intern-Project/BUDGET SALES ANALYSIS/Data/Product.xlsx"

sales_df = pd.read_excel(sales_path)
product_df = pd.read_excel(product_path)

# Merge Sales and Product datasets on ProductKey
merged_data = pd.merge(sales_df, product_df, on='ProductKey', how='inner')

# Group by Color and calculate total SalesAmount
color_sales = merged_data.groupby('Color')['SalesAmount'].sum().reset_index()

# Sort by SalesAmount
color_sales_sorted = color_sales.sort_values(by='SalesAmount', ascending=False)

# Display the results
print("Sales by Color:")
print(color_sales_sorted)

# Plot sales trends by Color
plt.figure(figsize=(10, 6))
plt.bar(color_sales_sorted['Color'], color_sales_sorted['SalesAmount'], color='skyblue')
plt.title('Sales by Color')
plt.xlabel('Color')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()

# Group by ProductLine and calculate total SalesAmount
product_line_sales = merged_data.groupby('ProductLine')['SalesAmount'].sum().reset_index()

# Sort by SalesAmount
product_line_sales_sorted = product_line_sales.sort_values(by='SalesAmount', ascending=False)

# Display the results
print("\nSales by ProductLine:")
print(product_line_sales_sorted)

# Plot sales trends by ProductLine
plt.figure(figsize=(10, 6))
plt.bar(product_line_sales_sorted['ProductLine'], product_line_sales_sorted['SalesAmount'], color='lightgreen')
plt.title('Sales by ProductLine')
plt.xlabel('Product Line')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```
